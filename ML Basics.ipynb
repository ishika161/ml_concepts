# Working with numpy

```python
import numpy as np
```

```python
# Creation of numpy arrays
# note:unlike arrays, numpy arrays cannot have heterogenous data; elementwise operations are allowed; contiguously stored
list1=[10,20,30]
print("concatenation: ",list1*2)
ar1=np.array([10,20,30])
print("element wise operation: ",ar1*2)
```

```python
# nparray converts all data to single type
ar2=np.array([1,-2,3.5,'%'])
print("Converted to string U32:",ar2)
```

```python
# attributes of nparray
print("ndim- number of dimension: ",ar1.ndim)
print("shape- size of array of each dimension: ",ar1.shape)
print("size- total number of elements: ",ar1.size)
print("dtype- data type of elements of aray: ",ar1.dtype)
print("itemsize- size in bytes of each element: ",ar1.itemsize)
```

```python
# Creation of numpy arrays
#1 for a particular datatype
a1=np.array([1,2,3,4],dtype='float32')
print(a1)
# 2 for zeroes
a2=np.zeros((2,4))
print(a2)
# 3 for ones
a3=np.ones((2,4))
print(a3)
# 4 for a range
a4=np.arange(1,6,2)
print(a4)
```

```python
# reshaping arrays
a5=np.ones((2,4))
print(a5)
a6=a5.reshape((4,2))
print(a6)
a7=a5.reshape((1,8))
print(a7)
```

## Data strucutres in pandas - series and data frames
series- 1D array

data frames- 2D labelled data structure.

## Series 

```python
# Creation of series
# 1. From scalar values

s1=pd.Series([1,2,3,4,5])
print("Series 1 :",s1)

# 2. from numpy array

ar1=np.array([1,2,3,4,5])
s2=pd.Series(ar1)
print("Series 2: ",s2)

#3. From dictionary

d1={1:"one",2:"two",3:"three"}
s3=pd.Series(d1)
print("Series 3:",s3)
```

```python
# Accessing Elements of Series
# Indexing
print("Indexing",s1[0])

#Slicing
print("Slicing",s1[0:])
```

###  Attributes of a series
###  name, index.name, values, size, empty
###  Methods of a series
###  head, count, tail

```python
#Operations on Series
# 1. Addition - direct , .add()

print(s1+s2)
print(s1.add(s2,fill_value=0))
```

```python
# subtraction - direct , sub()
print(s1+s2)
# replace nan with 100
print(s1.add(s2,fill_value=100))
```

## DataFrame

```python
# Creation of Data Frame
# 1. Empty data Frame
d1=pd.DataFrame()
print("first",d1)

#2. From numpy ndarrays
a1=np.array([1,2,3,4])
a2=np.array([5,6,7,8])
a3=np.array([9,10,11,12])
d2=pd.DataFrame([a1,a2,a3])
print("second",d2)

#3. from series
s11=pd.Series([1,2,3,4],index=['a','b','c','d'])
s12=pd.Series([5,6,7,8],index=['e','f','g','h'])
s13=pd.Series([9,10,11,12],index=['i','j','k','l'])
d3=pd.DataFrame([s11,s12,s13])
print("Third",d3)

#4. from list of dictionaries
li_dict=[{1:"one",2:"two"},{3:"three",4:"four"}]
d4=pd.DataFrame(li_dict)
print("Fourth",d4)

#5. from dictionary of lists
dict_li={"one":[1,2,3],"two":[4,5,6],"three":[7,8,9]}
```

# Adding rows and columns to dataframe

```python
# demo df
result={'preeti':pd.Series([50,40,60],index=['Maths','Science','GK']),
'kabir':pd.Series([90,98,88],index=['Maths','Science','GK']),
'dimple':pd.Series([75,80,85],index=['Maths','Science','GK']),
'vikram':pd.Series([98,88,99],index=['Maths','Science','GK']),
'reet':pd.Series([99,98,97],index=['Maths','Science','GK']),
'rohan':pd.Series([45,50,58],index=['Maths','Science','GK'])}
df=pd.DataFrame(result)
print(df)
```

```python
# adding column
df['prem']=[80,85,75]
print(df)
```

```python
# setting a column with all same values
df['rahul']=90
print(df)
```

```python
# adding row
df.loc['MI']=[78,88,76,86,87,77,97,99]
print(df)
```

```python
# droping a row - axis=0
# droping a column - axis=1

df=df.drop('rahul',axis=1)
print(df)
df=df.drop('MI',axis=0)
print(df)
```

```python
# renaming column
df=df.rename({'prem':'rahul'},axis='columns')
print(df)
```

```python
# renaming row
df=df.rename({'GK':'English'},axis='index')
print(df)
```

```python
# Accessing data through indexing
# for a row use .loc
# for cloumn use name directly
print("for row: ",df.loc['Maths'])
print("for column: ",df.loc[:,'preeti'])
print("for column: ",df['kabir'])
```

```python
# Accessing individual element
print((df.loc['Maths'])['preeti'])

# using .at : use(label,colname)
print(df.at['Maths','preeti'])

# using .iat : use(index,colname)
```

```python
# conditonal accessing
# for students who score more than 90 in maths
print(df.loc['Maths']>90)

# for rahul , subjects in which he scored more than 80
print(df['rahul']>80)
```

```python
# accessing marks of preeti in Maths  ".loc(row,col)"
print(df.loc['Maths','preeti'])
```

```python
# filtering rows "false row is omitted"
print(df)
print(df.loc[[True,False,True]])
```

```python
# join , merge , append dfs
se1=pd.Series([1,2,3,4],index=['a','b','c','d'])
se2=pd.Series([5,6,7],index=['e','f','g'])

df2=pd.DataFrame([se1,se2])
print(df2)
```

```python
df2=df2.append(df,sort='False')
print(df2)
```
